## This is a file with examples of using the search engine system

```python
"""Import relevent modules"""
from Searcher.Searcher import *
```

```python
"""Setup input file paths"""
w2v_path = "wiki-news-300d-1M.vec"
database_path = "database.csv"
```

```python
"""Load the word vector model"""
w2v = load_w2v(w2v_path)
```

```python
"""Initiate the searcher object, which will automatically crawl the database provided"""
searcher = Searcher(w2v=w2v, database=database_path)
```

```python
"""similarity searching"""
startup_description = "new start up aiming at low income customers, dedicated in green energy"
```

```python
# based on crawled data
%time searcher.update_similarity(startup_description, col=2)
```

```python
# based on summarized data
%time searcher.update_similarity(startup_description, col=3)
```
